```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all" # enable multiple output in one cell
```

```python
fig2a_df = pd.read_csv("D:/Projects/BW Mathematics/fig2a.csv")

dim_ai5 = fig2a_df.loc[(fig2a_df['Metric']=='AI')&(fig2a_df['Dimension']==5),'Delta']
dim_ai10 = fig2a_df.loc[(fig2a_df['Metric']=='AI')&(fig2a_df['Dimension']==10),'Delta']
dim_ai20 = fig2a_df.loc[(fig2a_df['Metric']=='AI')&(fig2a_df['Dimension']==20),'Delta']
dim_ai30 = fig2a_df.loc[(fig2a_df['Metric']=='AI')&(fig2a_df['Dimension']==30),'Delta']
dim_ai50 = fig2a_df.loc[(fig2a_df['Metric']=='AI')&(fig2a_df['Dimension']==50),'Delta']
dim_ai100 = fig2a_df.loc[(fig2a_df['Metric']=='AI')&(fig2a_df['Dimension']==100),'Delta']

dim_bw5 = fig2a_df.loc[(fig2a_df['Metric']=='BW')&(fig2a_df['Dimension']==5),'Delta']
dim_bw10 = fig2a_df.loc[(fig2a_df['Metric']=='BW')&(fig2a_df['Dimension']==10),'Delta']
dim_bw20 = fig2a_df.loc[(fig2a_df['Metric']=='BW')&(fig2a_df['Dimension']==20),'Delta']
dim_bw30 = fig2a_df.loc[(fig2a_df['Metric']=='BW')&(fig2a_df['Dimension']==30),'Delta']
dim_bw50 = fig2a_df.loc[(fig2a_df['Metric']=='BW')&(fig2a_df['Dimension']==50),'Delta']
dim_bw100 = fig2a_df.loc[(fig2a_df['Metric']=='BW')&(fig2a_df['Dimension']==100),'Delta']

fig2b_df = pd.read_csv("D:/Projects/BW Mathematics/fig2b.csv")

prop_ai10 = fig2b_df.loc[(fig2b_df['Metric']=='AI')&(fig2b_df['Proportion']==10),'Delta']
prop_ai20 = fig2b_df.loc[(fig2b_df['Metric']=='AI')&(fig2b_df['Proportion']==20),'Delta']
prop_ai40 = fig2b_df.loc[(fig2b_df['Metric']=='AI')&(fig2b_df['Proportion']==40),'Delta']
prop_ai60 = fig2b_df.loc[(fig2b_df['Metric']=='AI')&(fig2b_df['Proportion']==60),'Delta']
prop_ai80 = fig2b_df.loc[(fig2b_df['Metric']=='AI')&(fig2b_df['Proportion']==80),'Delta']

prop_bw10 = fig2b_df.loc[(fig2b_df['Metric']=='BW')&(fig2b_df['Proportion']==10),'Delta']
prop_bw20 = fig2b_df.loc[(fig2b_df['Metric']=='BW')&(fig2b_df['Proportion']==20),'Delta']
prop_bw40 = fig2b_df.loc[(fig2b_df['Metric']=='BW')&(fig2b_df['Proportion']==40),'Delta']
prop_bw60 = fig2b_df.loc[(fig2b_df['Metric']=='BW')&(fig2b_df['Proportion']==60),'Delta']
prop_bw80 = fig2b_df.loc[(fig2b_df['Metric']=='BW')&(fig2b_df['Proportion']==80),'Delta']
```

```python
fig, axes = plt.subplots(2, 2, figsize=(9, 6))

# Set theme
for ax_row in axes:
    for ax in ax_row:
        # background
        ax.set_facecolor('#F8FAFC')
        fig.patch.set_facecolor('white') # A light gray
        # grid
        ax.grid(True, which='major', axis='both', linestyle='-', linewidth=0.7, color='lightgray', alpha=0.8, zorder=0)
        ax.grid(True, which='minor', axis='both', linestyle=':', linewidth=0.4, color='lightgray', alpha=0.5, zorder=0)
        ax.minorticks_on() # Ensure minor ticks are on if you want minor gridlines
        # borders
        ax.spines['top'].set_visible(True)
        ax.spines['right'].set_visible(True)
        ax.spines['left'].set_visible(True)
        ax.spines['bottom'].set_visible(True)
        ax.spines['top'].set_color('black')
        ax.spines['right'].set_color('black')
        ax.spines['left'].set_color('black')
        ax.spines['bottom'].set_color('black')
        ax.spines['top'].set_linewidth(0.8)
        ax.spines['right'].set_linewidth(0.8)
        ax.spines['left'].set_linewidth(0.8)
        ax.spines['bottom'].set_linewidth(0.8)
        # ticks
        ax.tick_params(which='major', axis='both', length=3, color='black')
        ax.tick_params(which='minor', axis='both', length=1.5, color='black')
        # Keep tick label font size
        ax.tick_params(axis='x', labelsize=11)
        ax.tick_params(axis='y', labelsize=11)
        # Set y-label (if desired)
        ax.set_ylabel('Density', fontsize=14)

sns.kdeplot(dim_ai5, fill=True, label='5', ax=axes[0,0], zorder=2)
sns.kdeplot(dim_ai10, fill=True, label='10', ax=axes[0,0], zorder=2)
sns.kdeplot(dim_ai20, fill=True, label='20', ax=axes[0,0], zorder=2)
sns.kdeplot(dim_ai30, fill=True, label='30', ax=axes[0,0], zorder=2)
sns.kdeplot(dim_ai50, fill=True, label='50', ax=axes[0,0], zorder=2)
sns.kdeplot(dim_ai100, fill=True, label='100', ax=axes[0,0], zorder=2)
axes[0,0].set_xlabel(r"$\Delta_{ij}^{AI}$", fontsize=14)
axes[0,0].legend(loc='upper right', title='Dimension', fontsize=11, title_fontsize=11)
axes[0,0].set_title('(A)', loc='left', fontsize=14, ha='left', y=1.02)

sns.kdeplot(dim_bw5, fill=True, label='5', ax=axes[0,1], zorder=2)
sns.kdeplot(dim_bw10, fill=True, label='10', ax=axes[0,1], zorder=2)
sns.kdeplot(dim_bw20, fill=True, label='20', ax=axes[0,1], zorder=2)
sns.kdeplot(dim_bw30, fill=True, label='30', ax=axes[0,1], zorder=2)
sns.kdeplot(dim_bw50, fill=True, label='50', ax=axes[0,1], zorder=2)
sns.kdeplot(dim_bw100, fill=True, label='100', ax=axes[0,1], zorder=2)
axes[0,1].set_xlabel(r"$\Delta_{ij}^{BW}$", fontsize=14)
axes[0,1].legend(loc='upper right', title='Dimension', fontsize=11, title_fontsize=11)

sns.kdeplot(prop_ai10, fill=True, label='10%', ax=axes[1,0], zorder=2)
sns.kdeplot(prop_ai20, fill=True, label='20%', ax=axes[1,0], zorder=2)
sns.kdeplot(prop_ai40, fill=True, label='40%', ax=axes[1,0], zorder=2)
sns.kdeplot(prop_ai60, fill=True, label='60%', ax=axes[1,0], zorder=2)
sns.kdeplot(prop_ai80, fill=True, label='80%', ax=axes[1,0], zorder=2)
axes[1,0].set_xlabel(r"$\Delta_{ij}^{AI}$", fontsize=14)
axes[1,0].legend(loc='upper right', title='Proportion', fontsize=11, title_fontsize=11)
axes[1,0].set_title('(B)', loc='left', fontsize=14, ha='left', y=1.02)

sns.kdeplot(prop_bw10, fill=True, label='10%', ax=axes[1,1], zorder=2)
sns.kdeplot(prop_bw20, fill=True, label='20%', ax=axes[1,1], zorder=2)
sns.kdeplot(prop_bw40, fill=True, label='40%', ax=axes[1,1], zorder=2)
sns.kdeplot(prop_bw60, fill=True, label='60%', ax=axes[1,1], zorder=2)
sns.kdeplot(prop_bw80, fill=True, label='80%', ax=axes[1,1], zorder=2)
axes[1,1].set_xlabel(r"$\Delta_{ij}^{BW}$", fontsize=14)
axes[1,1].legend(loc='upper left', title='Proportion', fontsize=11, title_fontsize=11)

plt.tight_layout()
plt.savefig("D:/Projects/BW Mathematics/Figure 2 New.jpg", dpi=300, bbox_inches='tight')
plt.show()
```
