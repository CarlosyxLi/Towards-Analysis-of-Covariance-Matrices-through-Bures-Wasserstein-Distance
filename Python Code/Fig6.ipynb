```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all" # enable multiple output in one cell
```

```python
fig6a_df = pd.read_csv("D:/Projects/BW Mathematics/fig6a.csv")
fig6b1_df = pd.read_csv("D:/Projects/BW Mathematics/fig6b1.csv")
fig6b2_df = pd.read_csv("D:/Projects/BW Mathematics/fig6b2.csv")
fig6b3_df = pd.read_csv("D:/Projects/BW Mathematics/fig6b3.csv")
fig6c1_df = pd.read_csv("D:/Projects/BW Mathematics/fig6c1.csv")
fig6c2_df = pd.read_csv("D:/Projects/BW Mathematics/fig6c2.csv")
fig6c3_df = pd.read_csv("D:/Projects/BW Mathematics/fig6c3.csv")

a1 = fig6a_df.loc[(fig6a_df['Algorithm']=='Project Mean Alogrithm'),'SSD']
a2 = fig6a_df.loc[(fig6a_df['Algorithm']=='Inductive Mean Alogrithm'),'SSD']
a3 = fig6a_df.loc[(fig6a_df['Algorithm']=='Cheap Mean Alogrithm'),'SSD']

b1_num5 = fig6b1_df.loc[fig6b1_df['Num']==5,'RelativeSSD']
b1_num10 = fig6b1_df.loc[fig6b1_df['Num']==10,'RelativeSSD']
b1_num20 = fig6b1_df.loc[fig6b1_df['Num']==20,'RelativeSSD']
b1_num50 = fig6b1_df.loc[fig6b1_df['Num']==50,'RelativeSSD']
b1_num100 = fig6b1_df.loc[fig6b1_df['Num']==100,'RelativeSSD']

b2_num5 = fig6b2_df.loc[fig6b2_df['Num']==5,'RelativeSSD']
b2_num10 = fig6b2_df.loc[fig6b2_df['Num']==10,'RelativeSSD']
b2_num20 = fig6b2_df.loc[fig6b2_df['Num']==20,'RelativeSSD']
b2_num50 = fig6b2_df.loc[fig6b2_df['Num']==50,'RelativeSSD']
b2_num100 = fig6b2_df.loc[fig6b2_df['Num']==100,'RelativeSSD']

b3_num5 = fig6b3_df.loc[fig6b3_df['Num']==5,'RelativeSSD']
b3_num10 = fig6b3_df.loc[fig6b3_df['Num']==10,'RelativeSSD']
b3_num20 = fig6b3_df.loc[fig6b3_df['Num']==20,'RelativeSSD']
b3_num50 = fig6b3_df.loc[fig6b3_df['Num']==50,'RelativeSSD']
b3_num100 = fig6b3_df.loc[fig6b3_df['Num']==100,'RelativeSSD']

c1_num5 = fig6c1_df.loc[fig6c1_df['Num']==5,'RelativeSSD']
c1_num10 = fig6c1_df.loc[fig6c1_df['Num']==10,'RelativeSSD']
c1_num20 = fig6c1_df.loc[fig6c1_df['Num']==20,'RelativeSSD']
c1_num50 = fig6c1_df.loc[fig6c1_df['Num']==50,'RelativeSSD']
c1_num100 = fig6c1_df.loc[fig6c1_df['Num']==100,'RelativeSSD']

c2_num5 = fig6c2_df.loc[fig6c2_df['Num']==5,'RelativeSSD']
c2_num10 = fig6c2_df.loc[fig6c2_df['Num']==10,'RelativeSSD']
c2_num20 = fig6c2_df.loc[fig6c2_df['Num']==20,'RelativeSSD']
c2_num50 = fig6c2_df.loc[fig6c2_df['Num']==50,'RelativeSSD']
c2_num100 = fig6c2_df.loc[fig6c2_df['Num']==100,'RelativeSSD']

c3_num5 = fig6c3_df.loc[fig6c3_df['Num']==5,'RelativeSSD']
c3_num10 = fig6c3_df.loc[fig6c3_df['Num']==10,'RelativeSSD']
c3_num20 = fig6c3_df.loc[fig6c3_df['Num']==20,'RelativeSSD']
c3_num50 = fig6c3_df.loc[fig6c3_df['Num']==50,'RelativeSSD']
c3_num100 = fig6c3_df.loc[fig6c3_df['Num']==100,'RelativeSSD']
```

```python
fig = plt.figure(figsize=(9,5))
gs_all = fig.add_gridspec(nrows=1, ncols=2, width_ratios=[1,3], wspace=0.15, hspace=0.3)
gs_bc = gs_all[:, 1].subgridspec(nrows=2, ncols=3, wspace=0.25, hspace=0.3)

# A
ax_a = fig.add_subplot(gs_all[0])
sns.kdeplot(a1, fill=True, label='Project Mean Alogrithm', ax=ax_a, zorder=2)
sns.kdeplot(a2, fill=True, label='Inductive Mean Alogrithm', ax=ax_a, zorder=2)
sns.kdeplot(a3, fill=True, label='Cheap Mean Alogrithm', ax=ax_a, zorder=2)
ax_a.set_xlabel(r"(A) $SSD^{\circ}$", fontsize=14)
ax_a.set_ylabel('Density', fontsize=14)
ax_a.legend(loc='upper center', bbox_to_anchor=(0.5, -0.12), ncol=1, frameon=False, fontsize=11)

# B1
ax_b1 = fig.add_subplot(gs_bc[0,0])
sns.kdeplot(b1_num5, fill=True, label='5', ax=ax_b1, zorder=2)
sns.kdeplot(b1_num10, fill=True, label='10', ax=ax_b1, zorder=2)
sns.kdeplot(b1_num20, fill=True, label='20', ax=ax_b1, zorder=2)
sns.kdeplot(b1_num50, fill=True, label='50', ax=ax_b1, zorder=2)
sns.kdeplot(b1_num100, fill=True, label='100', ax=ax_b1, zorder=2)
ax_b1.set_xlabel('')
ax_b1.set_ylabel('Density', fontsize=14)
ax_b1.set_xticks([0,1,2,3])
ax_b1.set_yticks([0,1,2,3,4,5])
ax_b1.text(0.98, 0.97, "Dim = 5", fontsize=12, transform=ax_b1.transAxes, horizontalalignment='right', verticalalignment='top')
# B2
ax_b2 = fig.add_subplot(gs_bc[0,1])
sns.kdeplot(b2_num5, fill=True, label='5', ax=ax_b2, zorder=2)
sns.kdeplot(b2_num10, fill=True, label='10', ax=ax_b2, zorder=2)
sns.kdeplot(b2_num20, fill=True, label='20', ax=ax_b2, zorder=2)
sns.kdeplot(b2_num50, fill=True, label='50', ax=ax_b2, zorder=2)
sns.kdeplot(b2_num100, fill=True, label='100', ax=ax_b2, zorder=2)
ax_b2.set_xlabel(r"(B) $SSD^{I}-SSD^{P}$", labelpad=1, fontsize=14)
ax_b2.set_ylabel('')
ax_b2.set_yticks([0,0.4,0.8,1.2,1.6])
ax_b2.text(0.98, 0.97, "Dim = 20", fontsize=12, transform=ax_b2.transAxes, horizontalalignment='right', verticalalignment='top')
# B3
ax_b3 = fig.add_subplot(gs_bc[0,2])
sns.kdeplot(b3_num5, fill=True, label='5', ax=ax_b3, zorder=2)
sns.kdeplot(b3_num10, fill=True, label='10', ax=ax_b3, zorder=2)
sns.kdeplot(b3_num20, fill=True, label='20', ax=ax_b3, zorder=2)
sns.kdeplot(b3_num50, fill=True, label='50', ax=ax_b3, zorder=2)
sns.kdeplot(b3_num100, fill=True, label='100', ax=ax_b3, zorder=2)
ax_b3.set_xlabel('')
ax_b3.set_ylabel('')
ax_b3.set_yticks([0,0.3,0.6,0.9,1.2])
ax_b3.text(0.98, 0.97, "Dim = 50", fontsize=12, transform=ax_b3.transAxes, horizontalalignment='right', verticalalignment='top')

# C1
ax_c1 = fig.add_subplot(gs_bc[1,0])
sns.kdeplot(c1_num5, fill=True, label='5', ax=ax_c1, zorder=2)
sns.kdeplot(c1_num10, fill=True, label='10', ax=ax_c1, zorder=2)
sns.kdeplot(c1_num20, fill=True, label='20', ax=ax_c1, zorder=2)
sns.kdeplot(c1_num50, fill=True, label='50', ax=ax_c1, zorder=2)
sns.kdeplot(c1_num100, fill=True, label='100', ax=ax_c1, zorder=2)
ax_c1.set_xlabel('')
ax_c1.set_ylabel('Density', fontsize=14)
ax_c1.text(0.98, 0.97, "Dim = 5", fontsize=12, transform=ax_c1.transAxes, horizontalalignment='right', verticalalignment='top')
# C1
ax_c2 = fig.add_subplot(gs_bc[1,1])
sns.kdeplot(c2_num5, fill=True, label='5', ax=ax_c2, zorder=2)
sns.kdeplot(c2_num10, fill=True, label='10', ax=ax_c2, zorder=2)
sns.kdeplot(c2_num20, fill=True, label='20', ax=ax_c2, zorder=2)
sns.kdeplot(c2_num50, fill=True, label='50', ax=ax_c2, zorder=2)
sns.kdeplot(c2_num100, fill=True, label='100', ax=ax_c2, zorder=2)
ax_c2.set_xlabel(r"(C) $SSD^{C}-SSD^{P}$", labelpad=1, fontsize=14)
ax_c2.set_ylabel('')
ax_c2.set_xticks([0,10,20,30])
ax_c2.text(0.98, 0.97, "Dim = 20", fontsize=12, transform=ax_c2.transAxes, horizontalalignment='right', verticalalignment='top')
# C2
ax_c3 = fig.add_subplot(gs_bc[1,2])
sns.kdeplot(c3_num5, fill=True, label='5', ax=ax_c3, zorder=2)
sns.kdeplot(c3_num10, fill=True, label='10', ax=ax_c3, zorder=2)
sns.kdeplot(c3_num20, fill=True, label='20', ax=ax_c3, zorder=2)
sns.kdeplot(c3_num50, fill=True, label='50', ax=ax_c3, zorder=2)
sns.kdeplot(c3_num100, fill=True, label='100', ax=ax_c3, zorder=2)
ax_c3.set_xlabel('')
ax_c3.set_ylabel('')
ax_c3.set_xticks([0,100,200,300])
ax_c3.set_yticks([0,0.2,0.4,0.6,0.8,1])
ax_c3.text(0.98, 0.97, "Dim = 50", fontsize=12, transform=ax_c3.transAxes, horizontalalignment='right', verticalalignment='top')

fig.subplots_adjust(left=0.05, right=0.95, bottom=0.1, top=0.9)

handles_c, labels_c = ax_c1.get_legend_handles_labels()
ax_c1.legend(handles_c, labels_c, 
             loc='upper center', 
             bbox_to_anchor=(1.7, -0.35), 
             title='Number of Matrices', 
             fontsize=11,
             title_fontsize=14,
             ncol=len(labels_c), 
             frameon=False)

for ax in fig.get_axes():
    # background
    ax.set_facecolor('#F8FAFC')
    fig.patch.set_facecolor('white') # A light gray
    # grid
    ax.grid(True, which='major', axis='both', linestyle='-', linewidth=0.7, color='lightgray', alpha=0.8, zorder=0)
    ax.grid(True, which='minor', axis='both', linestyle=':', linewidth=0.4, color='lightgray', alpha=0.5, zorder=0)
    ax.minorticks_on() # Ensure minor ticks are on if you want minor gridlines
    # borders
    ax.spines['top'].set_visible(True)
    ax.spines['right'].set_visible(True)
    ax.spines['left'].set_visible(True)
    ax.spines['bottom'].set_visible(True)
    ax.spines['top'].set_color('black')
    ax.spines['right'].set_color('black')
    ax.spines['left'].set_color('black')
    ax.spines['bottom'].set_color('black')
    ax.spines['top'].set_linewidth(0.8)
    ax.spines['right'].set_linewidth(0.8)
    ax.spines['left'].set_linewidth(0.8)
    ax.spines['bottom'].set_linewidth(0.8)
    # ticks
    ax.tick_params(axis='both', which='major', length=3, color='black')
    ax.tick_params(axis='both', which='minor', length=0, color='black')
    # Keep tick label font size
    ax.tick_params(axis='x', labelsize=11)
    ax.tick_params(axis='y', labelsize=11)

plt.savefig("D:/Projects/BW Mathematics/Figure 6 New.jpg", dpi=300, bbox_inches='tight')
plt.show()
```
