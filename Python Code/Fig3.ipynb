```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all" # enable multiple output in one cell
```

```python
fig3_df = pd.read_csv("D:/Projects/BW Mathematics/fig3.csv")

ai1 = fig3_df.loc[(fig3_df['Case']=='Case 1')&(fig3_df['Metric']=='AI'),'F_distance']
ai2 = fig3_df.loc[(fig3_df['Case']=='Case 2')&(fig3_df['Metric']=='AI'),'F_distance']
ai3 = fig3_df.loc[(fig3_df['Case']=='Case 3')&(fig3_df['Metric']=='AI'),'F_distance']

bw1 = fig3_df.loc[(fig3_df['Case']=='Case 1')&(fig3_df['Metric']=='BW'),'F_distance']
bw2 = fig3_df.loc[(fig3_df['Case']=='Case 2')&(fig3_df['Metric']=='BW'),'F_distance']
bw3 = fig3_df.loc[(fig3_df['Case']=='Case 3')&(fig3_df['Metric']=='BW'),'F_distance']
```

```python
fig, axes = plt.subplots(1, 3, figsize=(9, 2))

# plt.rcParams['text.usetex'] = True
# plt.rcParams['font.family'] = 'serif'
# plt.rcParams['font.serif'] = ['Computer Modern Roman']

# Set theme
for ax in axes:
    # background
    ax.set_facecolor('#F8FAFC')
    fig.patch.set_facecolor('white') # A light gray
    # grid
    ax.grid(True, which='major', axis='both', linestyle='-', linewidth=0.7, color='lightgray', alpha=0.8, zorder=0)
    ax.grid(True, which='minor', axis='both', linestyle=':', linewidth=0.4, color='lightgray', alpha=0.5, zorder=0)
    ax.minorticks_on() # Ensure minor ticks are on if you want minor gridlines
    # borders
    ax.spines['top'].set_visible(True)
    ax.spines['right'].set_visible(True)
    ax.spines['left'].set_visible(True)
    ax.spines['bottom'].set_visible(True)
    ax.spines['top'].set_color('black')
    ax.spines['right'].set_color('black')
    ax.spines['left'].set_color('black')
    ax.spines['bottom'].set_color('black')
    ax.spines['top'].set_linewidth(0.8)
    ax.spines['right'].set_linewidth(0.8)
    ax.spines['left'].set_linewidth(0.8)
    ax.spines['bottom'].set_linewidth(0.8)
    # ticks
    ax.tick_params(axis='both', which='major', length=3, color='black')
    ax.tick_params(axis='both', which='minor', length=1.5, color='black')
    # Keep tick label font size
    ax.tick_params(axis='x', labelsize=11)
    ax.tick_params(axis='y', labelsize=11)

sns.kdeplot(ai1, fill=True, label=r"$d_{F}(M_{i}^{AI},\tilde{M}_{ij}^{AI})$", ax=axes[0], zorder=2)
sns.kdeplot(bw1, fill=True, label=r"$d_{F}(M_{i}^{BW},\tilde{M}_{ij}^{BW})$", ax=axes[0], zorder=2)
axes[0].set_ylim(0,31)
axes[0].set_xlabel('')
axes[0].set_ylabel('Density', fontsize=14)

sns.kdeplot(ai2, fill=True, label=r"$d_{F}(M_{i}^{AI},\tilde{M}_{ij}^{AI})$", ax=axes[1], zorder=2)
sns.kdeplot(bw2, fill=True, label=r"$d_{F}(M_{i}^{BW},\tilde{M}_{ij}^{BW})$", ax=axes[1], zorder=2)
axes[1].set_ylim(0,31)
axes[1].set_xlabel('')
axes[1].set_ylabel('')

sns.kdeplot(ai3, fill=True, label=r"$d_{F}(M_{i}^{AI},\tilde{M}_{ij}^{AI})$", ax=axes[2], zorder=2)
sns.kdeplot(bw3, fill=True, label=r"$d_{F}(M_{i}^{BW},\tilde{M}_{ij}^{BW})$", ax=axes[2], zorder=2)
axes[2].set_ylim(0,31)
axes[2].set_xlabel('')
axes[2].set_ylabel('')

handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, 
           loc='upper center', 
           bbox_to_anchor=(0.5, 0.12), 
           fontsize=14, 
           # title_fontsize=11,
           ncol=len(labels), 
           frameon=False, # Display legend box frame
           facecolor='white', # Match subplot background
           edgecolor='black', # Match spine color
           fancybox=False # Straight edges for theme_bw feel
          )

plt.tight_layout()
plt.savefig("D:/Projects/BW Mathematics/Figure 3 New.jpg", dpi=300, bbox_inches='tight')
plt.show()
```
