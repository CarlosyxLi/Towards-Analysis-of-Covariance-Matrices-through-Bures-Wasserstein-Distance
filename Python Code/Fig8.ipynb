```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter 

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all" # enable multiple output in one cell
```

```python
fig8_df = pd.read_csv("D:/Projects/BW Mathematics/fig8.csv")

x1 = fig8_df.loc[(fig8_df['metric']=='AI')&(fig8_df['subject']=='av'),'iteration']
y1 = fig8_df.loc[(fig8_df['metric']=='AI')&(fig8_df['subject']=='av'),'ssd']

x2 = fig8_df.loc[(fig8_df['metric']=='AI')&(fig8_df['subject']=='aw'),'iteration']
y2 = fig8_df.loc[(fig8_df['metric']=='AI')&(fig8_df['subject']=='aw'),'ssd']

x3 = fig8_df.loc[(fig8_df['metric']=='AI')&(fig8_df['subject']=='ay'),'iteration']
y3 = fig8_df.loc[(fig8_df['metric']=='AI')&(fig8_df['subject']=='ay'),'ssd']

x4 = fig8_df.loc[(fig8_df['metric']=='BW')&(fig8_df['subject']=='av'),'iteration']
y4 = fig8_df.loc[(fig8_df['metric']=='BW')&(fig8_df['subject']=='av'),'ssd']

x5 = fig8_df.loc[(fig8_df['metric']=='BW')&(fig8_df['subject']=='aw'),'iteration']
y5 = fig8_df.loc[(fig8_df['metric']=='BW')&(fig8_df['subject']=='aw'),'ssd']

x6 = fig8_df.loc[(fig8_df['metric']=='BW')&(fig8_df['subject']=='ay'),'iteration']
y6 = fig8_df.loc[(fig8_df['metric']=='BW')&(fig8_df['subject']=='ay'),'ssd']
```

```python
fig, axes = plt.subplots(2,3,figsize=(9,5))
plt.subplots_adjust(wspace=0.45, hspace=0.05)

# Set theme
for ax_row in axes:
    for ax in ax_row:
        # background
        ax.set_facecolor('#F8FAFC')
        fig.patch.set_facecolor('white') # A light gray
        # grid
        ax.grid(True, which='major', axis='both', linestyle='-', linewidth=0.7, color='lightgray', alpha=0.8, zorder=0)
        ax.grid(True, which='minor', axis='both', linestyle=':', linewidth=0.4, color='lightgray', alpha=0.5, zorder=0)
        ax.minorticks_on() # Ensure minor ticks are on if you want minor gridlines
        # borders
        ax.spines['top'].set_visible(True)
        ax.spines['right'].set_visible(True)
        ax.spines['left'].set_visible(True)
        ax.spines['bottom'].set_visible(True)
        ax.spines['top'].set_color('black')
        ax.spines['right'].set_color('black')
        ax.spines['left'].set_color('black')
        ax.spines['bottom'].set_color('black')
        ax.spines['top'].set_linewidth(0.8)
        ax.spines['right'].set_linewidth(0.8)
        ax.spines['left'].set_linewidth(0.8)
        ax.spines['bottom'].set_linewidth(0.8)
        # ticks
        ax.tick_params(which='major', axis='both', length=3, color='black')
        ax.tick_params(which='minor', axis='both', length=1.5, color='black')
        # Keep tick label font size
        ax.tick_params(axis='x', labelsize=12)
        ax.tick_params(axis='y', labelsize=12)

axes[0,0].plot(x1, y1, label='AI', color='steelblue', linewidth=1.5)
axes[0,0].set_title('av', fontsize=16)
axes[0,0].set_xticks([0,25,50,75])

axes[0,1].plot(x2, y2, label='AI', color='steelblue', linewidth=1.5)
axes[0,1].set_title('aw', fontsize=16)
axes[0,1].set_xticks([0,25,50,75])

axes[0,2].plot(x3, y3, label='AI', color='steelblue', linewidth=1.5)
axes[0,2].set_title('ay', fontsize=16)
axes[0,2].set_xticks([0,25,50,75])

axes[1,0].plot(x4, y4, label='BW', color='orange', linewidth=1.5)
axes[1,0].yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
axes[1,0].set_xticks([0,25,50,75])

axes[1,1].plot(x5, y5, label='BW', color='orange', linewidth=1.5)
axes[1,1].yaxis.set_major_formatter(FormatStrFormatter('%.1f'))
axes[1,1].set_xticks([0,25,50,75])

axes[1,2].plot(x6, y6, label='BW', color='orange', linewidth=1.5)
axes[1,2].yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
axes[1,2].set_xticks([0,25,50,75])

fig.supxlabel('Iteration Index', fontsize=14, x=0.55, y=0.04)
fig.supylabel('Sum of Squared Distances (SSD)', fontsize=14, ha='center', x=0.035)

from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], color='steelblue', lw=1.5, label='AI'),
    Line2D([0], [0], color='orange', lw=1.5, label='BW')
]
fig.legend(handles=legend_elements, 
           loc='lower center', 
           bbox_to_anchor=(0.55, -0.06), 
           fontsize=12, 
           title_fontsize=16,
           ncol=2, 
           frameon=False)

plt.tight_layout()
# plt.savefig("D:/Projects/BW Mathematics/Figure 8 New.jpg", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.ticker as mticker

def custom_formatter(x, pos):
    if x >= 10000 or x <= -10000:  # Apply comma for numbers 10,000 or more (positive or negative)
        return f'{int(x):,}'
    else:  # No comma for numbers less than 10,000 (absolute value)
        return f'{int(x)}'

formatter = mticker.FuncFormatter(custom_formatter)
```
